```python
import os 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains

import time 

options=webdriver.ChromeOptions()

options.add_experimental_option('prefs',{
    "download.default_directory": "/home/ankit/selenium-pdf/data",
    "download.prompt_for_download": False,
    "plugins.always_open_pdf_externally": True
})
download_directory="/home/ankit/selenium-pdf/data"
driver=webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()),options=options)
URL="https://bbnl.nic.in"
driver.get(URL)
driver.maximize_window()
driver.implicitly_wait(5)
count=1
read_more_button = driver.find_element(By.CSS_SELECTOR, ".coustom_services_readmore a[title='Know your Fiber - Read More...']")
read_more_button.click()

for p in range(1,37,1):
    try:
        read_more_button = driver.find_element(By.CSS_SELECTOR, ".coustom_services_readmore a[title='Know your Fiber - Read More...']")
        read_more_button.click()
    except NoSuchElementException:
        pass    
    area_element = driver.find_element(By.CSS_SELECTOR, f"map[name='FPMap1'] area[onmouseover=\"ddrivetip('{p}','#ffebcd', '200')\"]")
    st_name=area_element.get_attribute("title")
    print(st_name)
    try:
        driver.execute_script("arguments[0].scrollIntoView();", area_element)
        time.sleep(1)
        driver.execute_script("arguments[0].click();", area_element)
        print("Clicked successfully!")
    except Exception as e:
        print(f"Error clicking element: {e}")
        driver.save_screenshot("error_screenshot.png")
    select_element = driver.find_element(By.ID, "ddldist")
    select = Select(select_element)
    print(len(select.options))
    district=1
    for i in range(len(select.options)):
        select_element = driver.find_element(By.ID, "ddldist")
        select = Select(select_element)  
        option = select.options[i]
        dt_name=option.text 
        print(dt_name)
        print(option.get_attribute("value"))
        if option.get_attribute("value") == "-1":
            continue
        select.select_by_visible_text(option.text)
        block_links = driver.find_elements(By.CSS_SELECTOR, "td.Table_row_sub a")
        block=1
        for link in block_links:
            print(f"Downloading: {link.text}")
            link.click()
            time.sleep(3) 
            downloaded_files = sorted(os.listdir(download_directory), key=lambda x: os.path.getctime(os.path.join(download_directory, x)))
            latest_file = os.path.join(download_directory, downloaded_files[-1])  # Get the latest file in the directory
            new_file_name = f"C{count}_S{p}_{st_name}_D{district}_{dt_name}_B{block}_{link.text}.pdf"
            new_file_path = os.path.join(download_directory, new_file_name)

            # Ensure the file exists before renaming
            if os.path.isfile(latest_file):
                os.rename(latest_file, new_file_path)
                print(f"Renamed to: {new_file_name}")
            else:
                print(f"File not found for block: {link.text}")
            driver.implicitly_wait(5)   
            block=block+1
            count=count+1

        driver.back()
        try:
            area_element = driver.find_element(By.CSS_SELECTOR, f"map[name='FPMap1'] area[onmouseover=\"ddrivetip('{p}','#ffebcd', '200')\"]")
            try:
                driver.execute_script("arguments[0].scrollIntoView();", area_element)
                time.sleep(1)
                driver.execute_script("arguments[0].click();", area_element)
                print("Clicked successfully!")
            except Exception as e:
                print(f"Error clicking element: {e}")
                driver.save_screenshot("error_screenshot.png")
        except NoSuchElementException:
            pass
        time.sleep(2)
        district=district+1
    driver.back()
print("ankit")
```
